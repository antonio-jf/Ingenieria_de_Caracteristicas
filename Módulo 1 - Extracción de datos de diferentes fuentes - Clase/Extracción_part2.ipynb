```python
# Conteo de palabras. Expresiones regulares
import re
```

```python
re.split?
```

```python
# \w -> [a-z][A-Z][_][0-9]
# \W -> [*,.]
# \d -> [0-9]
# \s -> [\t][  ]
L = re.split('\W', texto)
L
```

```python
# Convertimos la lista a set
S = set(L)
S.discard('')
S
```

```python
# Buscamos las palabras no repetidas del set en la variable de texto para poderlas contar
re.findall?
```

```python
re.findall('en',texto, flags=re.I)
```

```python
d = {}
for palabra in S:
    d[palabra] = len(re.findall(palabra, texto, flags=re.I))
```

```python
d
```

```python
df_conteo_palabras = pd.DataFrame(d.items(), columns=['Palabra','Número'])
```

```python
df_conteo_palabras
```

```python
df_conteo_palabras = df_conteo_palabras.sort_values('Número', ascending = False)
```

```python
df_conteo_palabras
```

```python
df_conteo_palabras.sort_values('Número', ascending = False, inplace=True)
```

```python
df_conteo_palabras
```

```python
df_conteo_palabras.to_csv(ruta+'conteo_palabras.csv')
```

# Archivos XML

```python
import xml.etree.ElementTree as ET
```

```python
archivo_1 = ET.parse(ruta+'tabla_1.xml')
```

```python
raiz = archivo_1.getroot()
```

```python
for nodo in raiz:
    print(nodo.attrib, nodo.text, nodo.tag)
    for sn in nodo:
        print(sn.attrib, sn.text, sn.tag)
```

```python
archivo_2 = ET.parse(ruta+'tabla_2.xml')
root = archivo_2.getroot()
for nodo in root:
    print(nodo.tag, nodo.attrib, nodo.text)
```

```python
for nodo in root:
    for subn in nodo:
        print(subn.tag, subn.attrib, subn.text)
```

```python
#crear un diccionario que guarde el contenido del archivo xml
d={}
for nodo in raiz:
    d[nodo.tag] = []
for nodo in raiz:
    d[nodo.tag].append(nodo.attrib['name'])

for nodo in raiz:
    for sn in nodo:
        d[sn.tag]=[]

for nodo in raiz:
    for sn in nodo:
        d[sn.tag].append(sn.text)
```

```python
d
```

```python
df_tabla1 = pd.DataFrame(d)
df_tabla1
```

```python
df_tabla1.to_csv('Data/df_tabla2.csv')
```
