```python
# Conteo de palabras. Expresiones regulares
import re
```

```python
re.split?
```

```python
# \w -> [a-z][A-Z][_][0-9]
# \W -> [*,.]
# \d -> [0-9]
# \s -> [\t][  ]
L = re.split('\W', texto)
L
```

```python
# Convertimos la lista a set
S = set(L)
S.discard('')
S
```

```python
# Buscamos las palabras no repetidas del set en la variable de texto para poderlas contar
re.findall?
```

```python
re.findall('en',texto, flags=re.I)
```

```python
d = {}
for palabra in S:
    d[palabra] = len(re.findall(palabra, texto, flags=re.I))
```

```python
d
```

```python
df_conteo_palabras = pd.DataFrame(d.items(), columns=['Palabra','Número'])
```

```python
df_conteo_palabras
```

```python
df_conteo_palabras = df_conteo_palabras.sort_values('Número', ascending = False)
```

```python
df_conteo_palabras
```

```python
df_conteo_palabras.sort_values('Número', ascending = False, inplace=True)
```

```python
df_conteo_palabras
```

```python
df_conteo_palabras.to_csv(ruta+'conteo_palabras.csv')
```
